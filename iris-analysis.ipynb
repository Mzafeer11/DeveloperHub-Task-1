# Task 1: Exploring and Visualizing a Simple Dataset (Iris)

**Objective**: Learn how to load, inspect, and visualize a dataset to understand data trends and distributions.

### Step 1: Load the Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
iris = sns.load_dataset('iris')

# Print shape and column names
print(f"Dataset Shape: {iris.shape}")
print(f"Columns: {iris.columns.tolist()}")

# Show the first few rows
iris.head()
```

### Step 2: Summary Statistics
We use `.info()` to check for data types and missing values, and `.describe()` for numerical summary.

```python
print("--- Dataset Info ---")
iris.info()

print("\n--- Summary Statistics ---")
iris.describe()
```

### Step 3: Data Visualization

#### 3.1 Scatter Plot
Showing relationships between features (Sepal Length vs Sepal Width).

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', style='species', s=100)
plt.title('Relationship between Sepal Length and Width')
plt.show()
```

#### 3.2 Histograms
Showing distribution of values for each feature.

```python
iris.hist(edgecolor='black', linewidth=1.2, figsize=(12, 8))
plt.suptitle('Distribution of Iris Features')
plt.show()
```

#### 3.3 Box Plots
Identifying outliers and value ranges for each species.

```python
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.subplot(2, 2, 2)
sns.boxplot(x='species', y='sepal_width', data=iris)
plt.subplot(2, 2, 3)
sns.boxplot(x='species', y='petal_length', data=iris)
plt.subplot(2, 2, 4)
sns.boxplot(x='species', y='petal_width', data=iris)
plt.show()
```

### Step 4: Final Insights
The visualizations show that:
1. **Iris-Setosa** is easily distinguishable from others, especially by its petal length and width.
2. **Iris-Versicolor** and **Iris-Virginica** have some overlap but can be separated using petal features.
3. There are few outliers in **sepal width** for the Setosa species.
